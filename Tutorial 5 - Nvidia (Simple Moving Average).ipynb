## SIMPLE MOVING AVERAGES (SMA)

### <font color ="blue">BREAKDOWN OF THE COMMONLY USED MOVING AVERAGES (MA) AND THEIR CORRESPONDING TIME FRAMES

### 1. Short-Term Moving Averages:

20-day MA: This moving average represents the average price over the past 20 trading days and is used to capture short-term price movements.
50-day MA: This moving average represents the average price over the past 50 trading days and provides a slightly longer-term perspective.

### 2. Medium-Term Moving Averages:

This moving average represents the average price over the past 100 trading days and is often used to identify medium-term trends.

The 50-day MA provides a shorter-term perspective than the 100-day MA, capturing more recent price movements.

The 100-day MA represents a medium-term trend indicator.

The combination of these two moving averages can help identify potential entry and exit points based on short to medium-term trends.

### 3. Long-Term Moving Average:

200-day MA: This moving average represents the average price over the past 200 trading days and is widely regarded as a significant long-term trend indicator.

The 50-day MA provides a shorter-term perspective than the 200-day MA, capturing more recent price movements.

The 200-day MA represents a Long-term trend indicator.

The combination of these two moving averages can help identify potential entry and exit points based on short to Long-term trends.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import ta
import datetime
import warnings
warnings.filterwarnings("ignore")
```

```python
nvidia = yf.Ticker("NVDA")
```

```python
nvidia
```

```python
nvidia = nvidia.history(period="max")
```

```python
#Number of trading days
nvidia.shape
```

```python
#IPO launced
nvidia.index[0]
```

```python
nvidia['Close']
```

```python
plt.figure(figsize=(16,6))
nvidia['Close'].plot()
plt.grid(linestyle="--", color="grey", alpha=0.7)
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.title("NVDA")
plt.show()
plt.close()
```

```python
five_year = 365*5
five_year
```

```python
plt.figure(figsize=(16,6))
nvidia['Close'].tail(five_year).plot()
plt.grid(linestyle="--", color="grey", alpha=0.7)
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.title("NVDA")
plt.show()
plt.close()
```

```python
nvidia = nvidia.tail(five_year)
```

```python
nvidia.shape
```

```python
nvidia['SMA_20'] = nvidia['Close'].rolling(20).mean()
nvidia['SMA_50'] = nvidia['Close'].rolling(50).mean()
```

```python
nvidia.dropna()
```

```python
nvidia.shape
```

## SHORT TERM MOVING AVERAGE

```python
plt.figure(figsize=(16,6))
nvidia['Close'].plot()
nvidia['SMA_20'].plot(color="gold")
nvidia['SMA_50'].plot(color="red")
plt.grid(linestyle="--", color="grey", alpha=0.7)
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.title("NVDA")
plt.legend()
plt.show()
plt.close()
```

## MEDIUM TERM MOVING AVERAGE

```python
nvidia['SMA_100'] = nvidia['Close'].rolling(100).mean()
```

```python
nvidia.dropna(inplace=True)
```

```python
nvidia.shape
```

```python
plt.figure(figsize=(16,6))
nvidia['Close'].plot(linewidth=1)
nvidia['SMA_50'].plot(color="gold")
nvidia['SMA_100'].plot(color="red")
plt.grid(linestyle="--", color="grey", alpha=0.7)
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.title("NVDA")
plt.legend()
plt.show()
plt.close()
```

```python
golden_crosses = nvidia[nvidia.SMA_50 > nvidia.SMA_100]
bearish_crosses = nvidia[nvidia.SMA_50 < nvidia.SMA_100]
```

```python
golden_crosses.shape
```

```python
bearish_crosses.shape
```

```python
plt.figure(figsize=(16,6))
nvidia['Close'].plot(linewidth=1)
nvidia['SMA_50'].plot(color="gold")
nvidia['SMA_100'].plot(color="red")
plt.scatter(golden_crosses.index, golden_crosses.SMA_50, color="gold", marker="^", label="Goldern Cross")
plt.scatter(bearish_crosses.index, bearish_crosses.SMA_50 , color="red", marker="^", label="Bearish Cross")
plt.grid(linestyle="--", color="grey", alpha=0.7)
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.title("NVDA")
plt.legend()
plt.show()
plt.close()
```
