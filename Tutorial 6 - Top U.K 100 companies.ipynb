```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
class stock:
    def __init__(self, ticker):
        self.ticker = ticker
        
    def chart(self):
        return yf.download(self.ticker, period="1y")
```

```python
print(stock_ticker)
```

```python
united_kingdom_chart = stock(stock_ticker).chart()
```

## CONVERTING PENNY TO POUND 

```python
adj_close_values = united_kingdom_chart['Adj Close']/100
```

```python
adj_close_values[:5]
```

```python
adj_close_values.shape
```

```python
plt.figure(figsize=(15,6))
adj_close_values.max().plot()
adj_close_values.min().plot(color="red")
plt.title("MAXIMUM & MINIMUM FREQUENCY OF UK STOCK IN 1 YEAR", fontsize=16)
plt.xlabel("TOP 100 Companies", fontsize=16)
plt.ylabel("Price ()", fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.plot(adj_close_values.index, adj_close_values.values)
plt.title("UK 100 STOCK IN 1 YEAR", fontsize=16)
plt.xlabel("Date", fontsize=16)
plt.ylabel("Price ()", fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
mask_above_50 = adj_close_values > 50
adj_close_values[~mask_above_50] = np.nan
plt.figure(figsize=(15, 6))
plt.plot(adj_close_values.index, adj_close_values.values)
plt.title("UK 100 STOCK IN 1 YEAR", fontsize=16)
plt.xlabel("Date", fontsize=16)
plt.ylabel("Price ()", fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
selected_columns = adj_close_values.columns[adj_close_values.max() > 50]
selected_data_above_50 = adj_close_values[selected_columns]
```

#### This are list of ticker which price were above 50 pound in one year chart.

```python
selected_data_above_50.columns
```

```python
plt.figure(figsize=(15, 6))
for column in selected_data_above_50.columns:
    plt.plot(selected_data_above_50.index, selected_data_above_50[column], label=column)
plt.title("UK 100 STOCK IN 1 YEAR ABOVE 50 POUND", fontsize=16)
plt.xlabel("Date", fontsize=16)
plt.ylabel("Price ()", fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
